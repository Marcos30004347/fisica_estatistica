```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def logistic(r, x):
    return r * x * (1 - x)
```

```python
x = np.linspace(0, 1)
fig, ax = plt.subplots(1, 1)
ax.plot(x, logistic(2, x), 'k')
```

```python
def compute_and_plot_system(r, x0, n, ax = None):
    t = np.linspace(0, 1)

    ax.plot(t, logistic(r, t), 'k', lw=2)
    ax.plot([0, 1], [0, 1], 'k', lw=2)

    x = np.zeros(n)
    x[0] = x0
    
    for i in range(1, n):
        x[i] = logistic(r, x[i-1])

        ax.plot([x[i-1], x[i-1]], [x[i-1], x[i]], 'k', lw=1)
        ax.plot([x[i-1], x[i]], [x[i], x[i]], 'k', lw=1)
        ax.plot([x[i-1]], [x[i]], 'ok', ms=10, alpha=(i + 1) / n)

    fig, ax1 = plt.subplots(1, 1)
    ax1.plot(x)

    return x

def plot_given_r(r, x0, n):
    x = np.zeros((len(r), n))
    for i in range(len(r)):
        fig, ax = plt.subplots(1, 1)
        x[i] = compute_and_plot_system(r[i], x0[i], n, ax)

    return x
```

## 4

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 9))

n = 10000

r = np.linspace(1e-5, 4.0, n)

iterations = 1000

last = 100

x = 1e-5 * np.ones(n)

for i in range(iterations):
    x = logistic(r, x)
    if i >= (iterations - last):
        ax.plot(r, x, ',k', alpha=.25)

ax.set_xlim(1e-5, 4)
ax.set_title("Bifurcation diagram")
```

```python
plt.close('all')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(8, 9))

n = 10000

r = np.linspace(3.7, 3.9, n)

iterations = 1000

last = 100

x = 1e-5 * np.ones(n)

for i in range(iterations):
    x = logistic(r, x)
    if i >= (iterations - last):
        ax.plot(r, x, ',k', alpha=.25)

ax.set_xlim(3.7, 3.9)
ax.set_ylim(0.44, 0.56)
ax.set_title("Bifurcation diagram")
```
